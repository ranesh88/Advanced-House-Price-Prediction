```python
## We will be using logarithmic transformation

for feature in continuous_feature:
    data = df.copy()
    
    # Check for zeros or negative values
    if 0 in data[feature].unique():
        print(f"Skipping {feature} due to zeros or negative values.")
        continue  # Skip transformation for this feature
    
    # Apply logarithmic transformation
    data[feature] = np.log(data[feature])
    data['SalePrice'] = np.log(data['SalePrice'])
    
    # Plot histogram for the transformed feature
    plt.hist(data[feature], bins=30, alpha=0.7, color='blue', label=feature)
    plt.hist(data['SalePrice'], bins=30, alpha=0.7, color='orange', label='SalePrice')
    
    plt.xlabel('Log Transformed Values')
    plt.ylabel('Frequency')
    plt.title(f'Log Transformed Histogram: {feature} and SalePrice')
    plt.legend()
    plt.show()
```

### Feature Engineering

#### Missing Values

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing
features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

dataset=replace_cat_feature(df,features_nan)

dataset[features_nan].isnull().sum()
```

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values

for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=df[feature].median()
    
    ## create a new feature to capture nan values
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace=True)
    
df[numerical_with_nan].isnull().sum()
```

### Feature Scaling

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Create the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform your data
X_train_scaled = scaler.fit_transform(X_train)

# Convert the scaled data to a pandas DataFrame with original column names
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)

# View the first few rows
X_train_scaled_df.head()
```

```python
X_train_scaled_df.info()
```

### Feature Selection

```python
from sklearn.ensemble import RandomForestRegressor

# Fit a Random Forest model
model_rf = RandomForestRegressor()
model_rf.fit(X_train_scaled_df, y_train)

# Get feature importances
feature_importances = model_rf.feature_importances_

# Sort and get the top 10 most important features
important_features_idx = feature_importances.argsort()[-10:][::-1]
important_features = X_train_scaled_df.columns[important_features_idx]
print("Top 10 Important Features:", important_features)
```

```python
from sklearn.linear_model import Lasso

# Fit Lasso regression
lasso = Lasso(alpha=0.005)  # Alpha controls the regularization strength
lasso.fit(X_train_scaled_df, y_train)

# Get the selected features based on non-zero coefficients
selected_columns_lasso = X_train_scaled_df.columns[lasso.coef_ != 0]
print("Selected Features by Lasso:", selected_columns_lasso)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Fit RFE with a linear model (can be replaced with any other estimator)
model = LinearRegression()
rfe = RFE(estimator=model, n_features_to_select=10)  # Select top 10 features
X_train_scaled_df_rfe = rfe.fit_transform(X_train_scaled_df, y_train)

# Get the selected features' names
selected_columns_rfe = X_train_scaled_df.columns[rfe.support_]
print("Selected Features by RFE:", selected_columns_rfe)
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# Initialize the Lasso model for feature selection
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))  # Set the random_state to ensure reproducibility

# Fit the model to the scaled training data
feature_sel_model.fit(X_train_scaled_df, y_train)

# List of selected features
selected_feat = X_train_scaled_df.columns[(feature_sel_model.get_support())]

# Print total number of features, selected features, and features with coefficients shrunk to zero
print('Total features: {}'.format(X_train_scaled_df.shape[1]))
print('Selected features: {}'.format(len(selected_feat)))
print('Features with coefficients shrunk to zero: {}'.format(np.sum(feature_sel_model.estimator_.coef_ == 0)))

# Print selected features
print("Selected Features: ", selected_feat)
```

```python
from sklearn.feature_selection import SelectKBest, f_regression

# Select the top 10 features based on univariate feature selection
selector = SelectKBest(score_func=f_regression, k=10)  # k can be adjusted to the desired number of features
X_train_scaled_df_selected = selector.fit_transform(X_train_scaled_df, y_train)  # Assuming y_train is your target variable

# Get the selected features' names
selected_columns = X_train_scaled_df.columns[selector.get_support()]
print("Selected Features:", selected_columns)
```
